## Importing Library

```python
import pandas as pd
```

## Dataset Reading

```python
US_Accidents = pd.read_csv("US_Accidents_Dec19.csv")
```

```python
US_Accidents.columns
```

## Top 5 rows from datset

```python
US_Accidents.head(5)
```

```python
US_Accidents.shape
```

## Creating a new Column that will contains Year Information

```python
US_Accidents['Start_Year'] = pd.DatetimeIndex(US_Accidents['Start_Time']).year
US_Accidents['Month'] = pd.DatetimeIndex(US_Accidents['Start_Time']).month
```

## Unique Years from the dataset

```python
US_Accidents['Start_Year'].unique()
```

## Making a new dataframe on the base of year 2019

```python
US_Accidents_Updated = US_Accidents.loc[US_Accidents['Start_Year'].isin([2019])]
```

## Top Rows of New dataframe

```python
US_Accidents_Updated.head(5)
```

## Shape of the New Dataframe

```python
US_Accidents_Updated['Date']= pd.to_datetime(US_Accidents_Updated['Start_Time'])
US_Accidents_Updated.head()
```

```python
US_Accidents_Updated['AccidentDate'] = US_Accidents_Updated['Date'].dt.date
US_Accidents_Updated.head()
```

```python
US_Accidents_Updated.shape
```

# State Wise Accidents Count

```python
StateWiseAccidents = US_Accidents_Updated.groupby(['State', 'Severity']).State.agg('count').to_frame('StateCount').reset_index()
```

```python
StateWiseAccidents.to_csv("StateWise_Number_Accidents.csv", index=False)
```

# Traffic Signal Wise (True/False) Counts

```python
Traffic_Signal_Counts = US_Accidents_Updated.groupby(['Traffic_Signal']).Traffic_Signal.agg('count').to_frame('Count').reset_index()
Traffic_Signal_Counts['Traffic_Signal'] = Traffic_Signal_Counts['Traffic_Signal'].map({False: 'Accident Not On Traffic Signal', True: 'Accident On Traffic Signal'})
Traffic_Signal_Counts = Traffic_Signal_Counts.rename(columns={"Traffic_Signal": "Category"})
```

```python
Traffic_Signal_Counts.to_csv("accident_count_signal_donut_chart.csv", index=False)
```

# Weather Conditions Wise Accident Counts

```python
Weather_Conditions_Accidents = US_Accidents_Updated.groupby(['Weather_Condition']).Weather_Condition.agg('count').to_frame('Accident_Count').reset_index()

#get top 10 weather conditions
Weather_Conditions_Accidents = Weather_Conditions_Accidents.sort_values(by=['Accident_Count'], ascending=False)
Weather_Conditions_Accidents = Weather_Conditions_Accidents.head(10)
```

```python
Weather_Conditions_Accidents.to_csv("accident_count_Weather_Condition.csv", index=False)
```

# Month Wise Accident Counts

```python
Month_Wise_Accidents = US_Accidents_Updated.groupby(['AccidentDate', 'State']).Weather_Condition.agg('count').to_frame('Accident_Count').reset_index()
```

```python
Month_Wise_Accidents.to_csv("accident_count_month_wise.csv", index=False)
```
